```python
import requests
from bs4 import BeautifulSoup
import time
import csv
import random
import re
import pandas as pd
import numpy as np

with open('coaches.csv', 'r') as f:
    urls = pd.read_csv(f, dtype = {'year':str})
    
urls['team'] = ['St. Louis Cardinals (NFL)' if i == 'St. Louis Cardinals' else i for i in urls['team']]
urls['as'] = 'https://en.wikipedia.org/wiki/' + urls['year'] + ' ' + urls['team'] + ' season'
url_list = sorted([re.sub(' ','_',x) for x in urls['as'].unique()])

bad_urls = ['https://en.wikipedia.org/wiki/1968_Oakland_Raiders_season', \
            'https://en.wikipedia.org/wiki/1987_Washington_Redskins_season',\
           'https://en.wikipedia.org/wiki/1998_San_Francisco_49ers_season',\
           'https://en.wikipedia.org/wiki/1999_San_Francisco_49ers_season']

headers = {'User-Agent': \
           'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) \
           Chrome/58.0.3029.110 Safari/537.36'}

def assign_staff(list_of_folks, name_of_csv, team, year, indx):
    for idx in list_of_folks:
#         print(idx)
        staff = {'name':'','position':'','team': team,'year':year,'wikilink':''}
        name = ''
        position = ''
        wikilink = ''
        try:
            if re.search('—', idx):
                if (idx.find('—')>idx.find(' href') and re.search(' href',idx)):
                    position = re.findall('<i>(.*)</i>', idx)[0]
                    wikilink = re.findall(' href="(.*)" title="', idx)[0]
                    staff['wikilink'] = 'https://en.wikipedia.com' + wikilink
                    name = re.findall('">(.*)</a>', idx)[0]
                    
                else:
                    position = re.findall('(.*)—', idx)[0]

                    if re.search(' href',idx):
                        name = re.findall('">(.*)</a>', idx)[0]
                        wikilink = re.findall(' href="(.*)" title="', idx)[0]
                        staff['wikilink'] = 'https://en.wikipedia.com' + wikilink
                    else:
                        name = re.findall('—(.*)', idx)[0]
                    
            elif re.search('–', idx):
                if (idx.find('–')>idx.find(' href') and re.search(' href',idx)):
                    position = re.findall('<i>(.*)</i>', idx)[0]
                    wikilink = re.findall(' href="(.*)" title="', idx)[0]
                    staff['wikilink'] = 'https://en.wikipedia.com' + wikilink
                    name = re.findall('">(.*)</a>', idx)[0]
                    
                else:
                    position = re.findall('(.*)–', idx)[0]

                    if re.search(' href',idx):
                        name = re.findall('">(.*)</a>', idx)[0]
                        wikilink = re.findall(' href="(.*)" title="', idx)[0]
                        staff['wikilink'] = 'https://en.wikipedia.com' + wikilink
                    else:
                        name = re.findall('–(.*)', idx)[0]

            elif re.search('-', idx):
                position = re.findall('(.*)-', idx)[0]
                
                if re.search(' href',idx):
                    name = re.findall('">(.*)</a>', idx)[0]
                    wikilink = re.findall(' href="(.*)" title="', idx)[0]
                    staff['wikilink'] = 'https://en.wikipedia.com' + wikilink
                else:
                    name = re.findall('-(.*)', idx)[0]


            staff['name'] = name.strip()
            staff['position'] = position.strip()
            
        except Exception as e:
            print('Error inside', e, team, year, idx)
            
        if staff['name'] != '':
            name_of_csv.writerow(staff.values())
        else:
            continue
        

with open('nfl.csv', 'w', encoding='utf-8') as csvfile:
    nfl = csv.writer(csvfile)

    for indx, url in enumerate(url_list):
        block_of_text = ''
        if any([url == i for i in bad_urls]):
            continue
        try:
#             print(url)
            year = re.findall('https://en.wikipedia.org/wiki/(\d*)_',url)[0]
            team = re.sub('_',' ',re.findall('\d+_(.*)_season',url)[0])
            response = requests.get(url, headers = headers)
            text = BeautifulSoup(response.text, 'html.parser')
            
            if url == 'https://en.wikipedia.org/wiki/1976_Tampa_Bay_Buccaneers_season':
                block_of_text = re.findall('id="Coaching_staff_2"(.*)</tbody>', str(text), flags = re.DOTALL)
            
            elif re.search('id="Coaching_Staff"',str(text)):
                block_of_text = re.findall('id="Coaching_Staff"(.*)</tbody>', str(text), flags = re.DOTALL)

            elif re.search('id="Staff"',str(text)):
                block_of_text = re.findall('id="Staff"(.*)</tbody>', str(text), flags = re.DOTALL)

            elif re.search('id="Staff/Coaches"',str(text)):    
                block_of_text = re.findall('id="Staff/Coaches"(.*)</tbody>', str(text), flags = re.DOTALL)
                
            block_of_staff = re.split('</tbody>',block_of_text[0])[0]
            list_of_staff = re.findall('<li>(.*)</li>',block_of_staff)
            assign_staff(list_of_staff, nfl, team, year, indx)

            time.sleep(random.randint(1,3))
            
        except Exception as e:
            print(e, str(url))

```

```python
pos_abb = {'Asst ':'Assistant ', 'Asistant ':'Assistant ','Assistan ':'Assistant ','Assoc ':'Associate ',
           'AHC':'Assistant Head Coach','HC':'Head Coach','QC':'Quality Control','Coord ':'Coordinator ',
           '^Coach$':'Head Coach', '/Coach$':'/Head Coach', '^Coach/':'Head Coach/',
           'DC':'Defensive Coordinator', 'DL':'Dfensive Line', 'DB':'Defensive Backs', 'Def ':'Defensive ',
           'LB':'Linebackers', 'OLB':'Outside Linebackers', 'Dfensive ':'Defensive ',
           ' to the ':' ',
           'OC':'Offensive Coordinator', 'PGC':'Passing Game Coordinator', 'Pass Gm Coord':'Passing Game Coordinator',
           'RGC':'Run Game Coordinator', 'Off ':'Offensive ', 'OL':'Offensive Line', 'O-Line':'Offensive Line',
           'QB':'Quarterbacks', 'RB':'Running Backs', 'TE':'Tight Ends', 'WR':'Wide Receivers',
           
           'ST':'Special Teams', 'S&C':'Strength and Conditioning'}

name_dict = {'Correy Matthaei':'Corey Matthaei','Nate Kazcor':'Nate Kaczor','Derius Swinton$':'Derius Swinton II',
             'Ray Proschaska':'Ray Prochaska','Rod Humeniuk':'Rod Humenuik','Steve Matshall':'Steve Marshall',
             'Tom Breshnhan':'Tom Bresnahan','Tom Kanavu':'Tom Kanavy','Tom Mylinski':'Tom Myslinski'}

bad_names = ['Vacant']

front_office = ['Director','Executive','Ceo','Owner','President','Vp','Coo','Chairman','Officer','Founder','Partner',
                'Manager','Accounting','Accountant','Administrative','Controller','Counsel','Business','Operator',
                'Communications','Technology','Network','Relations','Marketing','Analyst',
                'Media','Photographer','Photography','Film','Tv','Editor','Video','Broadcasting', 'Strength',
                'Office','Contract','Secretary','Payroll','Ticket','Project','Treasurer',
                'Dentist','Surgeon','Surgery','Ophthalmologist','Physician','Nutrition','Science','Scientist',
                'Scout','Field','Personnel','On J','On F']

level_list = {'Coordinator':['Coordinator','Assistant Head','Assistant Coach','Special'],
              'Head':'Head',
              'Positional':['Receivers','Running Backs','Quarterbacks','Tight Ends','Offensive Line','Gaurds',
                            'Tackles','Defensive Backs','Backfield','Corners','Cornerbacks','Safeties','Front Seven',
                            'Nickel','Defensive Line','Secondary','Linebackers','Pass Rush','Kickers','Punters',
                            'Centers','Quality','Line','Backs','Line','Ends']}

phase_list = {'Head':'Head',
              'Offense':['Offense','Offensive','Receivers','Running Backs','Quarterbacks','Tight Ends','Tight',
                         'Running','Offensive Line','Gaurds','Centers'],
              'Defense':['Defensive','Defense','Backfield','Corners','Cornerbacks','Safeties','Front Seven',
                         'Nickel','Rush','Pass Rush','Defensive Line','Secondary','Linebackers','Pass Rush'],
              'Special Teams':['Special Teams','Kicking','Kickers','Punters','Punting','Return','Special']}

pfr = pd.read_csv('coaches.csv', dtype = {'year':str})

pfr.position = [re.sub('[.,:\'()]', '', i) for i in pfr.position]
pfr.position = [re.sub(' /','/', i) for i in pfr.position]
pfr.position = [re.sub('/ ',',', i) for i in pfr.position]

pfr.name = [re.sub('[’]','\'',i) for i in pfr.name]
pfr.name = [re.sub('Defense – |,|\.|performance – ','',i) for i in pfr.name]
pfr.name = [re.sub(' ','',i ,1) if re.search('^[A-Z] [A-Z]',i) else i for i in pfr.name]
            

wiki = pd.read_csv('nfl.csv',header=None)
wiki.columns=['name','position','team','year','wlink']
wiki.year = wiki.year.astype(str)
# wiki = wiki[(wiki.team != 'Oakland Raiders') | (wiki.year != '1968')] #not a table format

wiki = wiki.loc[(wiki.position != 'FFFFFF; text') & (wiki.position != 'FFFFFF;text')].reset_index()

wiki.drop(columns =['index'],inplace = True)

while any(re.search('[<]',wiki.name[i]) for i in range(len(wiki.name))):
    wiki.name = [re.findall('(.*)<',wiki.name[i])[0] if re.search('(.*)<',wiki.name[i]) else wiki.name[i] for i in range(len(wiki.name))]

while any(re.search('[<]',wiki.position[i]) for i in range(len(wiki.position))):
    wiki.position = [re.findall('(.*)<',wiki.position[i])[0] if \
                     re.search('(.*)<',wiki.position[i]) else wiki.position[i] for i in range(len(wiki.position))]
    
while any(re.search('[<]',wiki.wlink.astype(str)[i]) for i in range(len(wiki.wlink))):
    wiki.wlink = [re.findall('(.*)<',wiki.wlink.astype(str)[i])[0] if \
                  re.search('(.*)<',wiki.wlink.astype(str)[i]) else wiki.wlink[i] for i in range(len(wiki.wlink))]

wiki.position = [re.sub('[.,:\'()]', '', i) for i in wiki.position]
wiki.position = [re.sub(' /','/', i) for i in wiki.position]
wiki.position = [re.sub('/ ',',', i) for i in wiki.position]

wiki.name = [re.sub('[’]','\'',i) for i in wiki.name]
wiki.name = [re.sub('Defense – |,|\.|performance – ','',i) for i in wiki.name]
wiki.name = [re.sub(' ','',i ,1) if re.search('^[A-Z] [A-Z]',i) else i for i in wiki.name]



for i in list(pos_abb.keys()):

    for pos in range(0,len(pfr.position)):

        if re.search(i,pfr.position[pos]):
            pfr.loc[pos,'position'] = re.sub(i, pos_abb.get(i), pfr.loc[pos,'position'])
        else:
            continue

for i in list(pos_abb.keys()):
    
    for pos in range(len(wiki.position)):

        if re.search(i,wiki.position[pos]):
            wiki.loc[pos,'position'] = re.sub(i, pos_abb.get(i), wiki.loc[pos,'position'])
        else:
            continue

pfr.position = [str.title(i) for i in pfr.position]

wiki.position = [str.title(i) for i in wiki.position]


pfr = pfr.loc[[False if any([True if re.search(i, pfr.position[j]) else False for i in 
                       front_office]) else True for j in range(len(pfr.position))]].reset_index()

wiki = wiki.loc[[False if any([True if re.search(i, wiki.position[j]) else False for i in 
                       front_office]) else True for j in range(len(wiki.position))]].reset_index()

pfr.drop(columns =['index'],inplace = True)

wiki.drop(columns =['index'],inplace = True)


for i in list(name_dict.keys()):
    for nom in range(0,len(pfr.name)):
        if re.search(i,pfr.name[nom]):
            pfr.loc[nom,'name'] = re.sub(i, name_dict.get(i), pfr.loc[nom,'name'])
        else:
            continue

for i in list(name_dict.keys()):
    for nom in range(0,len(wiki.name)):
        if re.search(i,wiki.name[nom]):
            wiki.loc[nom,'name'] = re.sub(i, name_dict.get(i), wiki.loc[nom,'name'])
        else:
            continue


pfr = pfr.loc[[False if any([True if re.search(i, pfr.name[j]) else False for i in 
                       bad_names]) else True for j in range(len(pfr.name))]].reset_index()

wiki = wiki.loc[[False if any([True if re.search(i, wiki.name[j]) else False for i in 
                       bad_names]) else True for j in range(len(wiki.name))]].reset_index()

pfr.drop(columns =['index'],inplace = True)

wiki.drop(columns =['index'],inplace = True)


pfrclean = pd.DataFrame(pfr)
            
wikiclean = pd.DataFrame(wiki)

```

```python
wikiclean.to_csv('wikiclean.csv', encoding='utf-8', index=False)
pfrclean.to_csv('pfrclean.csv',encoding='utf-8', index=False)
```

```python
winloss=pfrclean[['team','year','win','loss','tie']]
winloss.drop_duplicates(inplace = True)
winloss.reset_index(inplace=True)
winloss.drop(columns = ['index'],inplace = True)
winloss = winloss.groupby(['team','year']).agg({'win':'sum','loss':'sum','tie':'sum'})
winloss.to_csv('winloss.csv',encoding='utf-8', index=False)
```

```python
wiki_unique = wikiclean[['name','year','team']].drop_duplicates().reset_index().drop(columns = ['index'])
wiki_unique['position']=['/'.join(wikiclean.position.loc[(wikiclean.name == wiki_unique.name[i]) & 
                                                     (wikiclean.year == wiki_unique.year[i]) &
                                                     (wikiclean.team == wiki_unique.team[i])].tolist()) 
                         for i in range(len(wiki_unique))]
wiki_unique.to_csv('wiki_unique.csv', encoding='utf-8', index=False)
```

```python
wiki_merge = pd.merge(pfrclean,wiki_unique,how='right',left_on=['team','year','name'],right_on=['team','year','name'],
                   suffixes = ('_p','_w'))
wiki_only = wiki_merge.loc[wiki_merge.isna().any(axis=1)].reset_index().filter(['name','team','year','position_w'],axis =1)
wiki_only.to_csv('wiki_only.csv', encoding='utf-8', index=False)
```

```python
combined = pd.merge(pfrclean,wiki_unique,how='outer',left_on=['team','year','name'],right_on=['team','year','name'],
                   suffixes = ('_p','_w'),)
combined = combined.fillna('')
combined['position']=combined['position_w']+'/'+combined['position_p']

combined.position = ['/'.join(set(str.split(combined.position[i].strip('^/').strip('/$'),'/'))) 
                     for i in range(len(combined.position))]

combined.drop(columns = ['position_p','position_w','link','win','loss','tie'],inplace = True)

combined['phase']=['' for i in combined.name]
combined['level']=['' for i in combined.name]

for i in range(len(combined)):
    for name, value in phase_list.items():
        if combined['phase'][i] =='':
            for word in re.split('[\W]',combined.position[i]):
                if word in value:
                    combined.loc[i,'phase'] = name
        else:
            continue

            
for i in range(len(combined)):
    for name, value in level_list.items():
        if combined['level'][i] =='':
            for word in re.split('[\W]',combined.position[i]):
                if word in value:
                    combined.loc[i,'level'] = name
        else:
            continue


combinedwl = pd.merge(combined,winloss,how='left',left_on=['team','year'],right_on=['team','year'])

combinedwl.drop_duplicates(inplace=True)
combinedwl.reset_index(inplace=True)
combinedwl.drop(columns = ['index'],inplace = True)
combinedwl.to_csv('combined.csv', encoding='utf-8', index=False)
```
